```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
# Read in M&M count data from HW2
counts_data = pd.read_csv(
    "https://docs.google.com/spreadsheets/d/e/2PACX-1vTdZfVH8hX4t0TGGaHQcShiV4Tk-OLj1xcD4GUeCz2M-OKfqXn-4O1593LfWdUsnr9Dun4CV5ivjbWi/pub?gid=469698708&single=true&output=csv"
)
counts_data = counts_data.iloc[:-1, :]
```

```python
# Plot the sampling distribution
counts_data["Totals"] = counts_data["Number_of_NonPrimary_Colored"] + counts_data["Number_of_Primary_Colored"]
counts_data["Proportions"] = counts_data["Number_of_Primary_Colored"] / counts_data["Totals"]
sns.histplot(data=counts_data, x="Proportions", bins=10)
```
